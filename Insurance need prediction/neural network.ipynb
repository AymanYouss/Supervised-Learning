```python
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from mlxtend.preprocessing import minmax_scaling
import pandas as pd
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, BatchNormalization, Conv2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')

print("Number of GPUs available is : ",len(physical_devices))
```

```python
df = pd.read_csv("Train.csv")
x = df.drop(columns=['target'])
y = df.target

for i in x.columns:
    x[i] = LabelEncoder().fit_transform(x[i])

x = minmax_scaling(x,columns=x.columns)

x
```

```python

train_data = tf.convert_to_tensor(x)
target_data = tf.convert_to_tensor(y)

model = Sequential([
    Dense(units=16, activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='relu')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(x,y,epochs=400,batch_size=100000,verbose=2,validation_split=0.2)
```
